```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
assignment1_csv = "/content/drive/MyDrive/phonology/assignment1/yourgoodpants.csv"
df = pd.read_csv(assignment1_csv, sep=',')

df.loc[df.Time_s == "--undefined--", "Time_s"] = None
df.loc[df.F0_Hz == "--undefined--", "F0_Hz"] = None
time = df.Time_s
f0 = df.F0_Hz

plt.plot(time, f0, "ro", label="pitch(Hz)")
plt.xlabel("time(sec)")
plt.ylabel("pitch(Hz)")
plt.legend(loc="best", ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
plt.close()
```

```python
assignment2_intensity_csv = "/content/drive/MyDrive/phonology/assignment2/yourgoodpants_intensity.csv"
df = pd.read_csv(assignment2_intensity_csv, sep=',')

df.loc[df.Time_s == "--undefined--", "Time_s"] = None
df.loc[df.Intensity_dB == "--undefined--", "Intensity_dB"] = None
time = df.Time_s
intensity = df.Intensity_dB

plt.plot(time, intensity, "ro", label="pitch(dB)")
plt.xlabel("time(sec)")
plt.ylabel("intensity(dB)")
plt.legend(loc="best", ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
plt.close()
```

```python
assignment2_formants_csv = "/content/drive/MyDrive/phonology/assignment2/yourgoodpants_formants.csv"
df = pd.read_csv(assignment2_formants_csv, sep=',')

df.loc[df.Time_s == "--undefined--", "Time_s"] = None
df.loc[df.F1_Hz == "--undefined--", "F1_Hz"] = None
df.loc[df.F2_Hz == "--undefined--", "F2_Hz"] = None
df.loc[df.F3_Hz == "--undefined--", "F3_Hz"] = None
df.loc[df.F4_Hz == "--undefined--", "F4_Hz"] = None
time = df.Time_s
f1 = df.F1_Hz.astype("float")
f2 = df.F2_Hz.astype("float")
f3 = df.F3_Hz.astype("float")
f4 = df.F4_Hz.astype("float")

plt.plot(time, f1, "ro", label="F1")
plt.plot(time, f2, "bo", label="F2")
plt.plot(time, f3, "go", label="F3")
plt.plot(time, f4, "yo", label="F4")
plt.xlabel("time(sec)")
plt.ylabel("frequency(Hz)")
plt.ylim([0, 7000])
plt.legend(loc="best", ncol=4, fontsize=14, frameon=True, shadow=True)
plt.show()
plt.close()
```
